```python
from bs4 import BeautifulSoup as bs
import requests
from tqdm.notebook import tqdm_notebook
import pandas as pd
```

```python
headers = {'authority': 'www.amazon.in',
           'accept-language': 'en-US,en;q=0.9',
           'referer': 'https://www.amazon.in/Indus-Valley-Cookware-Pre-Seasoned-Chapathi/product-reviews/B08YD1V5ZR/ref=cm_cr_dp_d_show_all_btm?ie=UTF8&reviewerType=all_reviews=8-3',
           'user-agent': 'Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/102.0.0.0 Safari/537.36'}
```

```python
def get_soup(url):
    r = requests.get(url, headers=headers,
    params={'url': url, 'wait': 2})
    soup = bs(r.text, 'html.parser')
    return soup
```

```python
reviewlist = []

def get_reviews(soup):
    reviews = soup.find_all('div', {'data-hook': 'review'})
    try:
        for item in reviews:
            review = {'Title': item.find('a', {'data-hook': 'review-title'}).text.strip(),
                      'Rating':float(item.find('i', {'data-hook': 'review-star-rating'}).text.replace('out of 5 stars', '').strip()),
                      'Review':item.find('span', {'data-hook': 'review-body'}).text.strip(),
                      'Review_Date':item.find('span', {'data-hook': 'review-date'}).text.replace('Reviewed in India ðŸ‡®ðŸ‡³ on', 'Reviewed in India on').strip()}
            reviewlist.append(review)
    except:
        pass
```

```python
for x in tqdm_notebook(range(160)):
    soup = get_soup(f'https://www.amazon.in/Indus-Valley-Cookware-Pre-Seasoned-Chapathi/product-reviews/B08YD1V5ZR/ref=cm_cr_dp_d_show_all_btm?ie=UTF8&reviewerType=all_reviews={x}')
    get_reviews(soup)
    if not soup.find('li', {'class': 'a-disabled a-last'}):
        pass
    else:
        break
```

```python
df = pd.DataFrame(reviewlist)
df
```

```python
df["Review_Date"] = df["Review_Date"].replace('Reviewed in India on','', regex=True)
```
